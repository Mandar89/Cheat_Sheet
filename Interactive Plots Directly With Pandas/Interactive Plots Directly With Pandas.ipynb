```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("C:/Users/mandar/Desktop/Nifty.csv", parse_dates=["Date"], index_col='Date')
```

```python
# resample/ aggregate the data by month-end
dataset_resample = dataset.resample(rule = 'M').mean()
```

```python
dataset_resample
```

```python
# Plotting with Pandas directly
```

```python
# Line Plot
dataset.plot(title = 'Nifty Index values in 2020', 
            xlabel = 'Values',
            figsize= (10,6));
```

```python
# Scatter Plot
dataset.plot(kind='scatter',
             x='NIFTY FMCG index',
            y='NIFTY Bank index',
            title='Scatter Plot for Nifty Index values in 2020',
            figsize=(10,6));
```

```python
# Histogram
dataset[['NIFTY FMCG index', 'NIFTY Bank index']].plot(kind='hist', figsize=(9,6), bins =30);
```

```python
# Bar Plot
dataset_resample.plot(kind='bar', figsize=(10,6));
```
